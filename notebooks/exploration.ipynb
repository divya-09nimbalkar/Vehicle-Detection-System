```python
import torch
import torchvision.transforms as transforms
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from src.evaluate import load_model
```

```python
# Load trained model
net = load_model("models/vehicle_resnet18.pth", num_classes=2)
```

```python
# Define transform (same as training)
transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
# Class labels
class_names = ["Vehicle", "NonVehicle"]
```

# 2. Helper Function

```python
def predict_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    # Apply transform
    from PIL import Image
    pil_img = Image.fromarray(img)
    input_tensor = transform(pil_img).unsqueeze(0)

    # Predict
    with torch.no_grad():
        outputs = net(input_tensor)
        _, pred = torch.max(outputs, 1)
        confidence = torch.softmax(outputs, dim=1)[0][pred].item()

    print(f"Prediction: {class_names[pred]} (Confidence: {confidence:.2f})")
```

# 3. Test on Your Dataset

```python
predict_image("data/processed/Vehicle/car1.png")
predict_image("data/processed/Vehicle/truck1.png")
predict_image("data/processed/NonVehicle/tree.png")
predict_image("data/processed/NonVehicle/building.png")
```
